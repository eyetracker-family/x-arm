```python
import numpy as np
center = 68
len = reachabilitystats.maxshape[0]
poseSolutionsIndexInSerialization = np.zeros([center*2+1,center*2+1,center*2+1,2])
reachabilitySet =np.zeros([len,4])
len
```

```python
for i in range(len):
    x = int(reachabilitystats[i,4]*100 + center)
    y = int(reachabilitystats[i,5]*100 + center)
    z = int(reachabilitystats[i,6]*100 + center)
    if poseSolutionsIndexInSerialization[x,y,z,0]<1:
        poseSolutionsIndexInSerialization[x,y,z,1] = i #start position in the reachability set of point(x,y,z) 
    poseSolutionsIndexInSerialization[x,y,z,0] += 1    #num of solutions
    reachabilitySet[i,0:4] = reachabilitystats[i,0:4]  #quaterion
    if i%10000 == 0:
        print i;
```
